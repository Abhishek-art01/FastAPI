```python
import os
import pdfplumber as pp
import pandas as pd

pdf_dir = r"D:\my_projects\air-india-data\data-dec-2025\DOC\FASTAG\pdf"
keyword = "idfc"  # part of the PDF name you want to read
csv_path = r"D:\my_projects\air-india-data\data-dec-2025\DOC\FASTAG\csv"

# Search for PDF containing the keyword (case-insensitive)
pdf_files = [f for f in os.listdir(pdf_dir)
             if f.lower().endswith(".pdf") and keyword.lower() in f.lower()]

if not pdf_files:
    raise FileNotFoundError(f"No PDF found containing '{keyword}'")

# If multiple matches, take the first one (or handle multiple)
pdf_path = os.path.join(pdf_dir, pdf_files[0])
print(f"Processing PDF: {pdf_files[0]}")

all_tables = []
pd.set_option("display.max_rows", None)
pd.set_option("display.max_colwidth", None)

with pp.open(pdf_path) as pdf:
    for page in pdf.pages:
        for table in page.extract_tables():
            if table:
                all_tables.append(pd.DataFrame(table))

df = pd.concat(all_tables, ignore_index=True)
df = df.drop(index=[0,1,2,3,4]).reset_index(drop=True)

# ------------------ Preview Markdown ------------------
chunk_size = 20

for i in range(0, len(df), chunk_size):
    print(df.iloc[i:i+chunk_size].to_markdown(index=True))
    print("\n" + "-" * 80 + "\n")
# ------------------ Save CSV with same PDF name ------------------
csv_name = os.path.splitext(pdf_files[0])[0] + ".csv"
csv_path = os.path.join(pdf_dir, csv_name)
df.to_csv(csv_path, index=False, encoding="utf-8-sig")
print(f"CSV saved as: {csv_path}")
```
